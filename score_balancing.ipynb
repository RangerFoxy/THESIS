```python
import pandas as pd
import statistics as st
```

```python
df_preprocessed = pd.read_json('./resources/preprocessed.json')
df_roberta = pd.read_json('./resources/roberta_sentiment_final.json')
df_flair = pd.read_json('./resources/flair_sentiment_with_score.json')
df_nltk = pd.read_json('./resources/nltk_sentiment.json')
df_textblob = pd.read_json('./resources/textblob_sentiment.json')
```

```python
type(df_flair['sentiment score'].values)
```

```python
df_flair[df_flair['sentiment score'].values <= 0.1]
```

```python
df_summed_score = df_flair
df_summed_score.head()
```

```python
df_summed_score = df_summed_score.drop('sentiment', 1)
df_summed_score
```

```python
df_summed_score['sentiment score'] = (df_roberta['sentiment score'].values + df_flair['sentiment score'].values + df_textblob['sentiment_score'].values + df_nltk['sentiment_score'].values) / 4
df_summed_score.head(50)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 10))

df_summed_score['sentiment score'] = scaler.fit_transform(df_summed_score['sentiment score'].values.reshape(-1,1))

df_summed_score
```

```python
summed_scores = df_summed_score.groupby('review index')['sentiment score'].mean()
```

```python
summed_scores
```

```python
df = summed_scores.to_frame().reset_index()
```

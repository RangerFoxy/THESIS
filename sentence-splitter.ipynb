```python
import pandas as pd
from nltk import tokenize
import nltk
nltk.download('punkt')
```

```python
df = pd.read_json('./resources/preprocessed.json')
df.info()
```

```python
review_index = 0
df_splitted = pd.DataFrame(columns=['review index', 'name', 'date', 'rating', 'review', 'thumbs up', 'total thumbs', 'thumbs down'])
```

```python
for index, row in df.iterrows():
    sentences = tokenize.sent_tokenize(row['review'])
    for sentence in sentences:
        df_splitted.loc[df_splitted.shape[0]] = [review_index, row['name'], row['date'], row['rating'], sentence, row['thumbs up'], row['total thumbs'], row['thumbs down']]
    review_index = review_index + 1    

df_splitted
```

```python
df_splitted.to_json('./resources/splitted.json')
```
